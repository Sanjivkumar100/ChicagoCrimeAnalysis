```python
import pandas as pd
crimedata=pd.read_excel('Crime_Data.xlsx')
```

```python
crimedata.info()
```

```python
crimedata['Location Description'].fillna('Unknown',inplace=True)
```

```python
crimedata.isnull().sum()
```

```python
blockcoordinates=crimedata.groupby('Block')[['Latitude','Longitude']].first()
crimedata['Latitude']=crimedata['Latitude'].fillna(crimedata['Block'].map(blockcoordinates['Latitude']))
crimedata['Longitude']=crimedata['Longitude'].fillna(crimedata['Block'].map(blockcoordinates['Longitude']))
```

```python
crimedata.drop(['Ward','Community Area'],axis=1,inplace=True)
```

```python
crimedata.isnull().sum()
```

```python
crimedata['Longitude']=crimedata['Longitude'].fillna('Unknown')
crimedata['Latitude']=crimedata['Latitude'].fillna('Unknown')
```

```python
crimedata.isnull().sum()
```

```python
# Split the datetime64[ns] column into separate Date and Time columns
crimedata['Case_Date'] = crimedata['Date'].dt.date  # Extract only the date part
crimedata['Case_Time'] = crimedata['Date'].dt.time  # Extract only the time part

# Verify the result
print(crimedata[['Date', 'Case_Date', 'Case_Time']].head())
```

```python
crimedata.drop('Date',axis=1,inplace=True)
```

```python
crimedata.drop('Only_Date',axis=1,inplace=True)
```

```python
crimedata.drop('Only_Time',axis=1,inplace=True)
```

```python
crimedata['Location'] = crimedata.apply(
    lambda row: f"({row['Latitude']}, {row['Longitude']})" if not pd.isna(row['Latitude']) and not pd.isna(row['Longitude']) else None,
    axis=1
)

crimedata.head()
```

```python
crimedata['Update_Date'] = crimedata['Updated On'].dt.date  # Extract only the date part
crimedata['Update_Time'] = crimedata['Updated On'].dt.time  # Extract only the time part

# Verify the result
print(crimedata[['Updated On', 'Update_Date', 'Update_Time']].head())
```

```python
crimedata.drop('Updated On',axis=1,inplace=True)
```

```python
crimedata.isnull().sum()
```

```python
crimedata.drop(['X Coordinate','Y Coordinate'],axis=1,inplace=True)
```

```python
cleanedata=crimedata[['ID','Case Number','Case_Date','Case_Time','Year','Block','IUCR','Primary Type','Description','Location Description','Arrest',	'Domestic','Beat','District','FBI Code','Update_Date','Update_Time','Latitude','Longitude','Location']]
```
